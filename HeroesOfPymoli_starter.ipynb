```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#player_count = int(len(purchase_data["SN"].unique()))

#total_player_count = pd.DataFrame({"Total Players": [player_count]})
#total_player_count
player_count = purchase_data["SN"].nunique()
total_player_count = pd.DataFrame({"Total Players": [player_count]})
total_player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Number of unique items
unique_items = len(purchase_data["Item ID"].unique())
#average price
average_price = purchase_data["Price"].mean()



#Number of Purchases 
purchases = len(purchase_data["Purchase ID"].unique())
#Total price 
total_price = purchase_data["Price"].sum()

total_purchasing_analysis= pd.DataFrame({"Number of Unique Items" : [unique_items], 
                                        "Average Price":[average_price],
                                        "Number of Purchases": [purchases],
                                        "Total Price": [total_price]})
total_purchasing_analysis["Average Price"]=total_purchasing_analysis["Average Price"].map("${:,.2f}".format)
total_purchasing_analysis["Total Price"]=total_purchasing_analysis["Total Price"].map("${:,.2f}".format)
 
total_purchasing_analysis.head()

```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python

#Gender Total
gender = pd.DataFrame(purchase_data.drop_duplicates(["SN"]))
gender_totals = gender.groupby(['Gender'])
#Creating a new dataframe with the gender percents and totals 
gender_df = pd.DataFrame({"Total Count":gender_totals["Gender"].count(),"Percentage of players":gender_totals["Gender"].count()/player_count*100})
gender_df["Percentage of players"] = gender_df["Percentage of players"].map("{:.2f}".format)
gender_df.sort_values("Total Count", ascending=False)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#gender purchasing analysis
gender_purchasing_analysis = pd.DataFrame(purchase_data)
purchase_by_gender=gender_purchasing_analysis.groupby(['Gender'])
purchase_by_gender_df = pd.DataFrame({"Purchase Count": purchase_by_gender["Price"].count(),
                                     "Average Purchase Price": purchase_by_gender["Price"].mean(),
                                     "Total Purchase Value": purchase_by_gender["Price"].sum(),
                                     "Average Total Purchase Per Person": purchase_by_gender["Price"].sum()/gender_df["Total Count"]})
purchase_by_gender_df["Average Purchase Price"]=purchase_by_gender_df["Average Purchase Price"].map("${:.2f}".format)
purchase_by_gender_df["Total Purchase Value"]=purchase_by_gender_df["Total Purchase Value"].map("${:.2f}".format)
purchase_by_gender_df["Average Total Purchase Per Person"] = purchase_by_gender_df["Average Total Purchase Per Person"].map("${:.2f}".format)
purchase_by_gender_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Age Bins
bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39","40+"]

#purchase_data["Age Ranges"] = pd.cut(player_count["Age"], bins, labels=group_names)
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins, right=False, labels=group_names)
purchase_data


#age_demographics.sort_index()
age_total = purchase_data["Age Ranges"].value_counts()
age_total
age_percent = (age_total / player_count * 100).round(2)
age_percent
age_demographics = pd.DataFrame({"Total Count": age_total, "Percentage of Players": age_percent})
age_demographics.sort_index()

```

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
age_purchasing_analysis = pd.DataFrame(purchase_data)
purchase_by_age=age_purchasing_analysis.groupby(["Age Ranges"])
purchase_by_age_df = pd.DataFrame({"Purchase Count": purchase_by_age["Price"].count(),
                                     "Average Purchase Price": purchase_by_age["Price"].mean(),
                                     "Total Purchase Value": purchase_by_age["Price"].sum(),
                                     "Average Total Purchase Per Person": purchase_by_age["Price"].sum()/age_demographics["Total Count"]})
purchase_by_age_df["Average Purchase Price"]=purchase_by_age_df["Average Purchase Price"].map("${:.2f}".format)
purchase_by_age_df["Total Purchase Value"]=purchase_by_age_df["Total Purchase Value"].map("${:.2f}".format)
purchase_by_age_df["Average Total Purchase Per Person"] = purchase_by_age_df["Average Total Purchase Per Person"].map("${:.2f}".format)
purchase_by_age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spenders= pd.DataFrame(purchase_data)
top_spenders=spenders.groupby(["SN"])
top_spenders_df = pd.DataFrame({"Purchase Count": top_spenders["Price"].count(),
                                     "Average Purchase Price":top_spenders["Price"].mean(),
                                     "Total Purchase Value": top_spenders["Price"].sum()
                                    })
top_spenders_df["Average Purchase Price"]=top_spenders_df["Average Purchase Price"].map("${:.2f}".format)
top_spenders_df["Total Purchase Value"]=top_spenders_df["Total Purchase Value"].map("${:.2f}".format)

top_spenders_df.sort_values("Total Purchase Value", ascending=False).head(10)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#popular_item_retrieval = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]
# Extract item Data
retrieval = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]

#Perform basic calculations
total_purchase_value = retrieval.groupby(["Item ID", "Item Name"]).sum()["Price"]
item_price = retrieval.groupby(["Item ID", "Item Name"]).mean()["Price"]
purchase_count = retrieval.groupby(["Item ID", "Item Name"]).count()["Price"]


retrieval_df= pd.DataFrame({"Purchase Count": purchase_count,"Item Price": item_price,
                             "Total Purchase Value": total_purchase_value, 
                             })
retrieval_df["Item Price"] = retrieval_df["Item Price"].map("${:,.2f}".format)
retrieval_df["Total Purchase Value"] = retrieval_df["Total Purchase Value"].map("${:,.2f}".format)


# Display the Item Table
retrieval_df.sort_values("Purchase Count", ascending=False).head(10)

```
